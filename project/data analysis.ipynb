```python
import pandas as pd
import numpy as np
import math as ma
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('csv/Tripadvisor_restaurant.csv', na_values="?")
df = df[~df.isna().any(axis=1)]
df
```

```python
y = df['rate']
y
```

```python
X = df['comment_num']
X
```

```python
X = sm.add_constant(X)
X
```

```python
model1 = sm.OLS(y, X).fit()
model1.summary()
```

```python
df.plot.scatter(x="rate", y="price_type", figsize=(15,10))
plt.plot(df.rate, model1.fittedvalues, color="r")
plt.show()
```
