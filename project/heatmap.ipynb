```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

# CSV
data = pd.read_csv('csv/location_price.csv', header=None, names=['Area', 'Value'], skiprows=1, encoding='utf-8')

with open('gis/Singapore.json', encoding='utf-8') as f:
    labelme_data = json.load(f)
regions = {}
for shape in labelme_data['shapes']:
    region_name = shape['label']
    points = shape['points']
    regions[region_name] = points


fig, ax = plt.subplots()


img = plt.imread('gis/Singapore.png')
img_height, img_width, _ = img.shape
ax.imshow(img, extent=[0, img_width, 0, img_height])


cmap = plt.cm.plasma
norm = plt.Normalize(vmin=data['Value'].min(), vmax=data['Value'].max())


for index, row in data.iterrows():
    region_name = row['Area']
    value = row['Value']
    if region_name in regions:
        points = np.array(regions[region_name])
        points[:, 1] = img_height - points[:, 1]
        color = cmap(norm(value))

        color_with_alpha = (*color[:3], 0.6)

        polygon = Polygon(points, closed=True, edgecolor='None', facecolor=color_with_alpha)
        ax.add_patch(polygon)

# 
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm._A = []
fig.colorbar(sm, ax=ax)

# 
plt.title('Average Price Heatmap')
plt.axis('off')
plt.show()


```

```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np

# CSV
data = pd.read_csv('csv/ideal_location.csv', header=None, names=['Area', 'Value'], skiprows=1,encoding='utf-8')

with open('gis/Singapore.json', encoding='utf-8') as f:
    labelme_data = json.load(f)
regions = {}
for shape in labelme_data['shapes']:
    region_name = shape['label']
    points = shape['points']
    regions[region_name] = points


fig, ax = plt.subplots()

# 
img = plt.imread('gis/Singapore.png')
ax.imshow(img, extent=[0, img.shape[1], 0, img.shape[0]])

# 
for index, row in data.iterrows():
    region_name = row['Area']
    value = row['Value']
    if region_name in regions:
        polygon = Polygon(regions[region_name], closed=True, edgecolor='None', facecolor=plt.cm.viridis(value*0.5))
        ax.add_patch(polygon)

# 
sm = plt.cm.ScalarMappable(cmap='viridis', norm=plt.Normalize(vmin=data['Value'].min(), vmax=data['Value'].max()))
sm._A = []
fig.colorbar(sm, ax=ax)

# 
plt.title('Region Heatmap')
plt.axis('off')
plt.show()
```
