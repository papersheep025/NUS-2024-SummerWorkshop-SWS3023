```python
def get_result():
    results = []
    content_element = driver.find_element(By.CSS_SELECTOR, "#content_left")

    #take note of the extra 's' in element
    search_results = content_element.find_elements(By.CSS_SELECTOR, '.result')
    print(len(search_results))
    
    for search_result in search_results:
        link_element = search_result.find_element(By.CSS_SELECTOR, "a")
        url = link_element.get_attribute("href")
        link_text = link_element.text
        description_text = None
        try:
            description_element = search_result.find_element(By.CSS_SELECTOR, "span[class^='content-right_']")
            description_text = description_element.text
        except:
            pass
        results.append({"url": url, "title": link_text, "description": description_text})
    return pd.DataFrame(results)
```

#### ❓Q1. Get the first 3 pages of search results and save it into a Pandas Dataframe named `df`
*Hint: you would need to do a sleep for a while before doing the scraping between pages:
`time.sleep(1)` will sleep for 1 seconds*

```python
import time
```

```python
curr_page = 1

overallDF = get_result()

while curr_page < 3:
    next_elem = driver.find_elements(By.CSS_SELECTOR, '#page > div > a.n')
    next_elem = next_elem[len(next_elem) - 1]
    next_elem.click()
    time.sleep(1)
        
    this_page_df = get_result()
    overallDF = pd.concat([overallDF, this_page_df], ignore_index=True)
    curr_page = curr_page + 1
    
overallDF
```
