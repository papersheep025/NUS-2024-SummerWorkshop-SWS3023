```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("weather.csv")
data
```

```python
#drop the last class column (play)
X = data.iloc[:,:-1]
X
```

```python
y = data["play"]
y
```

```python
y = pd.factorize(y)[0]
y
```

```python
X = pd.get_dummies(X)
X
```

# Naive Bayes

```python
from sklearn.naive_bayes import CategoricalNB, GaussianNB, MultinomialNB, BernoulliNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#as the dataset is all categorical data, we should use Categorical NB
nb_model1 = CategoricalNB()
nb_model1.fit(X_train, y_train)
y_pred1 = nb_model1.predict(X_test)
accuracy_score(y_test, y_pred1)
```

```python
#try GuassianNB (assume that the data is following normal distribution)
nb_model2 = GaussianNB()
nb_model2.fit(X_train, y_train)
y_pred2 = nb_model2.predict(X_test)
accuracy_score(y_test, y_pred2)
```

```python
#MultinomialNB is best for discrete values cases (e.g. word counts)
nb_model3 = MultinomialNB()
nb_model3.fit(X_train, y_train)
y_pred3 = nb_model3.predict(X_test)
accuracy_score(y_test, y_pred3)
```

```python
#BernoulliNB is for the case when the input is a set of boolean values
nb_model4 = BernoulliNB()
nb_model4.fit(X_train, y_train)
y_pred4 = nb_model4.predict(X_test)
accuracy_score(y_test, y_pred4)
```

# KNN

K-Nearest Neigbors is quite unique compared to other learning algorithm in that there isn't really a training phase. unlike most learning algorithm, in KNN the work happens during the testing phase.

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

# Support Vector Machines

```python
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
accuracy_score(y_test, y_pred_svm)
```

# Decision Tree

Scikit Learn does not handle visualization of decision tree for categorical columns nicely, so let's use another dataset

```python
from sklearn import tree
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df["species"] = data.target
df
```

```python
data.target_names
```

```python
X = df.iloc[:,0:-1]
X.columns
```

```python
y = df["species"]
y
```

```python
dt_model = tree.DecisionTreeClassifier(random_state=123)
dt_model.fit(X, y)

fig, axes = plt.subplots(figsize = (20,10), dpi=300)

#filled = True : paint nodes to indicate majority class for classification
tree.plot_tree(dt_model, feature_names=X.columns, class_names=data.target_names, filled=True)
plt.show()
```

# K-Fold Cross-Validation

```python
#we will use Iris Dataset
from sklearn.model_selection import cross_val_score
clf = SVC()

scores = cross_val_score(clf, X, y, cv=10)

print(scores)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
